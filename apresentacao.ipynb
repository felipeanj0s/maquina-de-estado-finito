## Introdução

O **protocolo TCP (Transmission Control Protocol)** é um dos principais protocolos do modelo TCP/IP, que é responsável por e assegurar e gerenciar conexões confiáveis entre dispositivos em uma rede.

Uma das formas de entender o funcionamento do TCP é por meio de uma **Máquina de Estados Finita (FSM)**, que modela as transições entre diferentes estados do protocolo.

## Estados do Protocolo TCP

O protocolo TCP é composto por diferentes estados que representam o ciclo de uma conexão. 
Os principais estados do Protocolo Controle de Transmissão são:

1. **CLOSED**: Representa uma conexão inativa.
2. **SYN_SENT**: Após enviar um SYN (Synchronyze) para iniciar a conexão, o cliente aguarda uma resposta do servidor.
3. **ESTABLISHED**: A conexão foi estabelecida, e dados podem ser trocados entre cliente e servidor.
4. **FIN_WAIT_1**: Após o cliente solicitar o encerramento, ele aguarda a confirmação do servidor.
5. **TIME_WAIT**: Após encerrar a conexão, o cliente aguarda um tempo para garantir que todos os pacotes tenham sido entregues.




### Diagrama Textual do TCP
```
[CLOSED] → open() 
→ [SYN_SENT] → receive() 
→ [ESTABLISHED] → close() 
→ [FIN_WAIT_1] → receive() 
→ [TIME_WAIT] → close() 
→ [CLOSED]
```
